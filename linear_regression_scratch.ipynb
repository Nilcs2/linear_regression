```python
# Task 1: Implement the mathematical formula for linear regression

import numpy as np

# generate synthetic data

rng = np.random.default_rng(42)

X = rng.random((100, 1))
y = 4 + 3 * X + rng.random((100, 1))

# Add bias term to feature matrix

X_b = np.c_[np.ones((100, 1)), X]

# Initialize the parameters.

theta = rng.random((2, 1))
learning_rate = 0.1
iterations = 1000

def predict(X, theta):
    return np.dot(X, theta)

# Task 2 : Use gradient descend to optimize the model parameters

def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    for _ in range(iterations):
        gradient = (1/m) * np.dot(X.T, (np.dot(X, theta) - y))
        theta -= learning_rate * gradient
    return theta

# Task 3 : Calculate Evaluation Matrix

def mean_squared_error(y_true, y_pred):
    return np.mean((y_pred - y_true)**2)

def r_squared(y_true, y_pred):
    ss_red = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - (ss_red/ss_tot)


# Perform the gradient descend

theta_optimized = gradient_descent(X_b, y, theta, learning_rate, iterations)

# Predictions and evaluation

y_pred = predict(X_b, theta_optimized)
mse = mean_squared_error(y, y_pred)
r2 = r_squared(y, y_pred)

print("Optimized parameters =", theta_optimized)
print("Mean squared error =", mse)
print("R squared =", r2)
```

```python
import matplotlib.pyplot as plt

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Synthetic Data')
plt.plot(X_b[:, 1], y_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.title('Gradient Descent Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
# Save the plot for your README
plt.savefig('regression_result.png')
print("Plot saved as regression_result.png")
```
